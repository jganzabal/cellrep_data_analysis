```python
import pandas as pd
from process_helper import plot_volcano, plot_donors_scatter
```

```python
cytokine = 'TNFa'
screen = 'CRISPRa'
screen_analysis_table = pd.read_csv(f'{cytokine}_screen_analysis_table.csv', index_col=False).iloc[:, 1:]
screen_analysis_table = screen_analysis_table[
    (screen_analysis_table['Screen'] == screen) & 
    (screen_analysis_table['Cytokine'] == cytokine) 
].set_index('Gene')
```

```python
top_n = 10
top_10_low = pd.read_csv('data_out/CRISPRa.TNFa.gene.low.txt', sep='\t')['group_id'][:top_n].to_list()
top_10_high = pd.read_csv('data_out/CRISPRa.TNFa.gene.high.txt', sep='\t')['group_id'][:top_n].to_list()
fig_1 = plot_donors_scatter(screen_analysis_table);
```
