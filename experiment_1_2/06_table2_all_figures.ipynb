```python
import pandas as pd
from process_helper import plot_volcano
from IPython.display import display, Markdown
```

# Load data and print table content

```python
df_g = pd.read_excel(
    '../data/Genome-wide-screens/data/science.abj4008_table_s2.xlsx', 
).groupby(['Screen_Version', 'CRISPRa_or_i', 'CD4_or_CD8', 'Cytokine'])
df_g.size().reset_index().rename(columns={0: 'total_rows'})
```

```python
for (study_type, screen, cell_type, cytokine), df_to_plot in df_g:
    display(Markdown('<div style="page-break-after: always; visibility: hidden"> \\pagebreak </div>'))
    display(Markdown(f'# {cytokine} in {cell_type}+, {screen} (Experiment: {study_type})'))
    
    f = plot_volcano(
        df_to_plot,
        x_label='LFC ' + '$'+f'{cytokine}'+'^{hi}/{cytokine}^{low}$',
        y_label='$-Log_{10}(FDR)$',
        plot_title=f'{study_type}-{screen}-{cell_type}-{cytokine}',
        fig_size=(12,10)
    );
    display(f)
    f.clf()
display(Markdown('<div style="page-break-after: always; visibility: hidden"> \\pagebreak </div>'))
```
