```python
import pandas as pd
from process_helper import plot_volcano, plot_donors_scatter
from IPython.display import display, Markdown
```

```python
df = pd.read_excel(
    '../data/Genome-wide-screens/data/science.abj4008_table_s2.xlsx', 
)
```

```python
df_g = df.groupby(['Screen_Version', 'CRISPRa_or_i', 'CD4_or_CD8', 'Cytokine'])
df_g.size()
```

```python
for (study_type, screen, cell_type, cytokine), df_to_plot in df_g:
    display(Markdown(f'### **Study**: {study_type}'))
    display(Markdown(f'### **Screen**: {screen}'))
    display(Markdown(f'### **Cell type**: {cell_type}'))
    display(Markdown(f'### **Cytokine**: {cytokine}'))
    f = plot_volcano(
        df_to_plot,
        x_label='LFC ' + '$'+f'{cytokine}'+'^{hi}/{cytokine}^{low}$',
        y_label='$-Log_{10}(FDR)$',
        plot_title=f'{study_type}-{screen}-{cell_type}-{cytokine}',
        fig_size=(10,5)
    );
    display(f)
    f.clf()
    print('#'*100)
```
